# **Hierarchy Clustering Templates**

## **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Impporting Datasets**

```python
dataset = pd.read_csv('Mall_Customers.csv')

X = dataset.iloc[:, [3, 4]].values
```

## **Using the Dendrogram to Find the Optimal Number of Clusters**

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Number of Clusters')
plt.ylabel('Euclidean Distances')
plt.show()
```

## **Training the Data using Optimal Number of Clusters**

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

## **Visualizing the Clusters**

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 75, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 75, c = 'green', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 75, c = 'red', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 75, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 75, c = 'magenta', label = 'Cluster 5')

plt.title('Cluster of Customrs')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Shopping Score (1-100')
plt.legend()
plt.show()
```
